### Analyse most commmon prompt types in chatgpt 

```python
df.head()
```

```python
def categorize_prompt(prompt):
    prompt = prompt.lower()

    categories = {
        "Information Retrieval": ["what", "who", "when", "where", "define", "explain"],
        "Summarization": ["summarize", "summary", "shorten"],
        "Translation": ["translate", "convert language"],
        "Code Generation": ["write code", "python code", "program", "script"],
        "Debugging": ["error", "bug", "fix", "debug"],
        "Data Analysis": ["analyze", "dataset", "dataframe", "statistics"],
        "Content Writing": ["blog", "article", "content", "write about"],
        "Creative Writing": ["story", "poem", "fiction", "creative"],
        "Brainstorming": ["ideas", "suggest", "brainstorm"],
        "Recommendation": ["best", "recommend", "top"],
        "Planning": ["plan", "schedule", "roadmap"],
        "Interview Preparation": ["interview", "questions", "mock interview"]
    }

    for category, keywords in categories.items():
        for keyword in keywords:
            if keyword in prompt:
                return category

    return "General Conversation"
```

```python
df['instruction']=df['instruction'].str.strip()
```

```python
df['prompt_type']=df['instruction'].apply(categorize_prompt)
```

```python
df['flesch_score'].mean()
```

```python
def readibility_level(s):
    if s>=90:
        return "very easy"
    elif s>=60:
        return "easy"
    elif s>=30:
        return "medium"
    elif s>=10:
        return "difficult"
    elif s>=0:
        return "very difficult"
    
    else:
        return "extrimly complicated"
```

```python
df['level']=df['flesch_score'].apply(readibility_level)
```

```python
df['level']
```

```python
df.columns
```

```python
fig = px.pie(
    df,
    names="level",
    title="Readability Level of ChatGPT Response"
)

fig.show()
```

### how verbose are chatgpt responses?

```python
df.columns
```

```python
def count_sentences(text):
    text=str(text)
    
    sentences=re.split(r'[.!?]',text)
    sentences=[s.strip() for s in sentences if s.strip()]
    return len(sentences)
```

```python
df['senteence_count']=df['output'].apply(count_sentences)
```

### Does additional context improve chatgpt response


```python
df.head()
```

```python
df['input']
```

```python
df['has_input']=df['input'].apply(lambda x: 0 if pd.isna(x) or str(x).strip()=='' else 1)
```

```python
df['has_input'].value_counts()
```

```python
df.columns
```

```python
comparison=df.groupby(['has_input'])[['output word count','flesch_score']].mean().reset_index()
```

```python
comparison
```

```python
comparison['Context']=comparison['has_input'].map({0:'No extra context',1:'extra context'})
```

```python
comparison
```

```python
px.bar(comparison,
       x='Context',
       y=['output word count','flesch_score'],
       barmode='group',
       title='effect of extra context on chatgpt responses',
       labels={
           "value":'Average Value',
           'Variable':'metric',
           'context':'prompt type'
       })
```
